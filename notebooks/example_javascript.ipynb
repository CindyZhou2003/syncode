```python
import sys, os
sys.path.append(os.getcwd() + '/../')

from syncode import Syncode
import warnings
warnings.filterwarnings('ignore')

model_name = "microsoft/phi-2"
```

```python
# Load the unconstrained original model
llm = Syncode(model = model_name, mode='original', max_new_tokens=250)
# Load the Syncode augmented model
syn_llm = Syncode(model = model_name, mode='grammar_mask', grammar='javascript', parse_output_only=True)
```

The standard LLM answer also includes a text part `A: You can use the following code:`. This text could be different everytime and hence it is difficult to extract the answer from the text. 

In addition, the LLM repeatedly outputs the same code.

Using SynCode solves this problem and the output is purely a single instance of the correct code!

```python
prompts = [
    
    "Write a JavaScript function named 'add' that adds two numbers.\n",

    "Instruct: Write a JavaScript function named 'add'.\nOutput:",

    "// Language: JavaScript\n// Task: Write a function named 'multiply' that multiplies two numbers.\n",

    "// Use JavaScript as the programming language to declare an array named 'names' with three strings.\n",

    "Define a JavaScript class named 'User' with a constructor.\n",

    "// Define a JavaScript class named 'Car' with a constructor that takes 'model' and 'year'.\n",

    "function factorial(n) {",

    "// Import the 'fs' module.\n",

    "Write HelloWorld.js, which prints Hello World!\n"
]

for i,p in enumerate(prompts):
    print(f"\nPrompt{i+1}: {p}\n")
    
    # 1. Standard (Unconstrained)
    output = llm.infer(p)[0]
    print(f"LLM output:\n{output}\n") 

    print("-"*30)
    print("\n")
    
    output = syn_llm.infer(p)[0]
    print(f"Syncocde LLM output:\n{output}\n")
    print("="*40)
```
